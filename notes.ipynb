# Requerimientos del punto B

### Limpieza y Normalización
El archivo de entrada es un CSV con los campos: Series Name, Season Number, Episode Number, Episode Title y Air Date.
* Series name: es obligatorio que este
* Season / Episode Number: Si flata, o es negativo o no es numero, se setea en 0
* Episode Title: Si falta o esta vacio, reemplazar por "Untitled Episode"
* Air Date: Si falta, está vacio o es una fecha invalida, reemplazar por "Unknown"
* Regla de descarte: Si faltan al mismo tiempo el numero de episodio, titulo y fecha se descarta el registro.
* Normalización: se dee limpiar espacios y pasar todo a minusculas.

### Duplicación de registros

Un episodio se considera **duplicado** si coincide en estas combinaciones de campos normalizados

1. (Serie, Temporada, Episodio)
2. (Serie, 0, Episodio, Titulo)
3. (Serie, Temporada, 0, Titulo)

Prioridad para elegir el mejor registro
* Se prefiere uno con **fecha valida** sobre el que diga "Unknown"
* Se prefiere uno con **titulo conocido** sobre "Unknown Episode"
* Se prefiere uno con **temporada y numero** validos
* En caso de empate, se queda con el **primero** que apareció
